```python
t1=tables[0].copy()
t1 = t1.drop(t1.index[[10,11]])
t1['Year'] = t1['Year'].astype(int)
t1['year']=t1['Year']+1
t1["gpd_percapita_change"]= t1["GDP per capita"].str.split(" ", n = 1, expand = True)[1].astype(float)
t1["ciw_change"]= t1["CIW"].str.split(" ", n = 1, expand = True)[1].astype(float)
t1=t1.drop(t1.columns[[0,1,2]], axis=1).set_index('year')
t1
```

```python
t0=tables[0].copy()
t0 = t0.drop(t0.index[11]).rename(columns={'Year': 'year'})
t0['year'] = t0['year'].astype(int)
t0["gpd_percapita_change"]= t0["GDP per capita"].str.split(" ", n = 1, expand = True)[0].astype(float)
t0["ciw_change"]= t0["CIW"].str.split(" ", n = 1, expand = True)[0].astype(float)
t0 = t0.drop(t0.columns[[1,2]], axis=1).set_index('year')
t0
```

```python
CIWvsGDP = pd.concat([t0, t1]). sort_index()
CIWvsGDP.tail(1)
```

# Get Feature Tables

```python
def table_maker_function(feature):
    url_a = f'https://uwaterloo.ca/canadian-index-wellbeing/reports/2016-canadian-index-wellbeing-national-report/{feature}'
    table_a = pd.read_html(url_a)[0].set_index('Year')
    url_b = f'https://uwaterloo.ca/canadian-index-wellbeing/reports/2016-canadian-index-wellbeing-national-report/{feature}-domain-raw-data'
    table_b = pd.read_html(url_b)[0].set_index('Year')
    Table = pd.concat([table_a,table_b], axis=1, sort=False)
    Table = Table.drop(Table.columns[[0,1]], axis=1)
    return Table
```

```python
education = table_maker_function('education')
education = education.rename(columns={'Education': 'education',
                                      'Percentage of children aged 0 to 5 years for whom there is a regulated centre-based childcare space': 'childcare',
                                      'Average amount of time spent in talk-based activities with children aged 0 to 14 years': 'talkbased_activity',
                                      'Average expenditure per public school student (2013$)': 'expenditure_perschool',
                                      'Ratio of students to educators in public schools': 'educators_perschool',
                                      'Average annual Canadian undergraduate tuition fees (2015$)': 'undergraduate_tuition', 
                                      'Percentage of Canadians 20 to 24 years of age in labour force completing high school': 'highschool_edu',
                                      'Percentage of 25 to 64 year olds in population with a university degree': 'university_edu', 
                                      'Percentage of adults aged 25 and older participating in education related activities': 'edu_activities',})
education.tail(1)
```

```python
healthy_populations = table_maker_function('healthy-populations')
healthy_populations = healthy_populations.rename(columns={'Healthy Populations': 'hlty_pop', 
                                                          'Life expectancy at birth in years': 'expectancy',
                                                          'Percentage of population that rates their overall health as very good or excellent': 'overall_health',
                                                          'Percentage of population that rates their mental health as very good or excellent': 'mental_health',
                                                          'Percentage of population with an absence of health or activity-based limitations': 'health_limitations',
                                                          'Percentage of daily or occasional smokers among teens aged 12 to 19 years': 'teen_smokers', 
                                                          'Percentage of population with self-reported diabetes': 'diabetes',
                                                          'Percentage of population getting influenza immunization in past year': 'influenza_immunization', 
                                                          'Percentage of Canadians with a regular medical doctor': 'family_doctor',})
healthy_populations.tail(1)
```
